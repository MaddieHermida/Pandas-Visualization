# Pandas Visualization

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\Cezar\Desktop\Bootcamp\Pandas\Panda Visualizations\Ice Cream Ratings.csv")
df.set_index('Date')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300)
```

```python
df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
df.plot.area(figsize = (10,3))
```

```python
df.plot.pie(y = 'Flavor Rating')
```
